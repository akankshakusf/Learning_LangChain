#### Ollama
Ollama supports embedding models, making it possible to build retrieval augmented generation (RAG) applications that combine text prompts with existing documents or other data.

```python
#get the embedding model from Ollama
from langchain_community.embeddings import OllamaEmbeddings
```

```python
#instantiate the model 
embedding= (
    OllamaEmbeddings(model="gemma2:2b")  # by default it uses llama 2
)
```

```python
embedding
```

```python
#passing 2 sentences to model to convert to vectors 
r1=embedding.embed_documents(
    [
        "Alpha is the first letter of the greek alphabet",
        "Beta is the second alphabet of the greek alphabet"
    ]
)
```

```python
r1[1]
```

```python
len(r1[0])
```

# now lets play model 

```python
embedding.embed_query("what is the second letter of the greek alphabet")
```
