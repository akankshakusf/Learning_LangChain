### Read about the model used "text-embedding-3-large" on the link -https://platform.openai.com/docs/models#embeddings

```python
#get the embedding model from Openai
from langchain_openai import OpenAIEmbeddings

#instantiate the model 
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
#pass some text to the model amd convert to vectors 
text="This is a tutorial on OpenAI Embedding"

#pass text to model
query_results=embeddings.embed_query(text)
query_results
```

### playing with speech.txt file 

```python
from langchain_community.document_loaders import TextLoader

#instantiate the loader
loader=TextLoader('speech.txt')
docs=loader.load()
docs
```

```python
#split the text per the recursive char splitter 
from langchain_text_splitters import RecursiveCharacterTextSplitter

#instantiate the splitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=20)
final_document=text_splitter.split_documents(docs)
final_document
```

```python
#instantiate the model 
embeddings_1024=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)

#pass text to model
query_results=embeddings_1024.embed_query(text)
query_results
```

### Here as we have list of documents now we will combine the Vector Embedding and Vector store

##### There are 3 type of vector store dbs we will use 1:
* FAISS
* CHROMA DB
* ASTRA DB


```python
#using chroma bd to store vectors
from langchain_community.vectorstores import Chroma

#initailize chroma db for documents 
db=Chroma.from_documents(final_document,embeddings_1024)
db
```

```python
### Retrieve the results from query vectorstore db
query="It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results=db.similarity_search(query)
print(retrieved_results)
```
